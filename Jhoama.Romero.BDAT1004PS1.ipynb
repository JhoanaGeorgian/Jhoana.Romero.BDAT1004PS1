```python
# Solution:
# This code was done using Visual Studio
using System;

class program
{
    //main code
    static bool inside(double x, double y, double x1,
                          double y1, double x2, double y2)
    {
        // put the conditions of the functions
        // we need that x1<x<x2 and y1<y<y2
        if (x1<x && x < x2 &&
            y1 < y && y < y2)
            return true;

        return false;
    }

    // Method
    public static void Main()
    {
        double x1, x2, y1, y2;
        //INPUT
        Console.WriteLine("Insert the x1 value of lower left corner: ");
        x1= Convert.ToDouble(Console.ReadLine());
        Console.WriteLine("Insert the y1 value of lower left corner: ");
        y1 = Convert.ToDouble(Console.ReadLine());
        Console.WriteLine("Insert the x2 value of lower right corner: ");
        x2 = Convert.ToDouble(Console.ReadLine());
        Console.WriteLine("Insert the y2 value of lower right corner: ");
        y2 = Convert.ToDouble(Console.ReadLine());
        // Fixed or given point 

        double x, y;
        Console.WriteLine("Insert your x point:  ");
        x = Convert.ToDouble(Console.ReadLine());
        Console.WriteLine("Insert your y point:  ");
        y = Convert.ToDouble(Console.ReadLine());
       
        // function call
        if (inside( x, y, x1,y1,  x2, y2))
            Console.Write("Does your given point (x,y) lie in the rectangle with lower left corner (x1, y1) and upper right corner (x2, y2)?: TRUE");
        else
            Console.Write("Does your given point (x,y) lie in the rectangle with lower left corner (x1, y1) and upper right corner (x2, y2)?: FALSE");
    }
}
```

# Question 6 Python
You can turn a word into pig-Latin using the following two rules (simplified):

- If the word starts with a consonant, move that letter to the end and append 'ay'. For example, 'happy' becomes 'appyhay' and 'pencil' becomes 'encilpay'.
- If the word starts with a vowel, simply append 'way' to the end of the word.

For example, 'enter' becomes 'enterway' and 'other' becomes 'otherway'. For
our purposes, there are 5 vowels: a, e, i, o, u (so we count y as a consonant).
Write a function pig() that takes a word (i.e., a string) as input and returns its pigLatin form. Your function should still work if the input word contains upper case
characters. Your output should always be lower case however.

**Source**:  a help source for this program was found in *StackOverflow* wep page. See the link:

[StackOverflow](https://stackoverflow.com/questions/23177250/converting-a-sentence-to-piglatin-in-python)

```python
def pig(x):
    vowels = list("aeiou")
    lowx=x.lower()
    if lowx[0] not in vowels:
        newx=lowx[1:] + lowx[0]+ 'ay'
    else:
        newx=lowx + 'way'
        
    return newx
   
```

```python
pig("pencil")
```

```python
pig('happy')
```

```python
pig('Enter')
```

# Question 10 Python

Encryption is the process of hiding the meaning of a text by substituting letters in the
message with other letters, according to some system. If the process is successful, no
one but the intended recipient can understand the encrypted message. Cryptanalysis
refers to attempts to undo the encryption, even if some details of the encryption are
unknown (for example, if an encrypted message has been intercepted). The first step
of cryptanalysis is often to build up a table of letter frequencies in the encrypted text.
Assume that the string letters is already defined as
*'abcdefghijklmnopqrstuvwxyz'*. Write a function called frequencies()
that takes a string as its only parameter, and returns a list of integers, showing the
number of times each character appears in the text. Your function may ignore any
characters that are not in letters.

```python
#define the function frequencies for our string as input
def frequencies(ourstring):

    #the letters defined in our alphabet
    defined = "abcdefghijklmnopqrstuvwxyz"
    
    #conver our input in a list of characters
    stringlist= ourstring.split()
    #Define a vector oz zeros of length of defined
    zer = [0]*len(defined)
    
    #iterates through the alphabet
    for i in range(0, len(defined)):
        
        for word in stringlist:
            
            if defined[i] in word:
                zer[i]=zer[i] + word.count(defined[i])
                
    return zer
```
